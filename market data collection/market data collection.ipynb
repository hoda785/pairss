```python
import yfinance as yf
from newsapi import NewsApiClient
import pandas as pd
import datetime as dt

# Replace with your own NewsAPI key
newsapi_key = 'cdbe46fb08ad469da4086f9b1038ea70'
newsapi = NewsApiClient(api_key=newsapi_key)

# Load the list of selected stocks
print("Loading selected stocks...")
selected_tickers_df = pd.read_csv("selected_stocks_financials.csv")
tickers = selected_tickers_df['symbol'].tolist()

# Function to fetch news sentiment data
def fetch_news_sentiment(ticker, start_date, end_date):
    print(f"Fetching news sentiment data for {ticker}...")
    news_data = newsapi.get_everything(q=ticker, from_param=start_date,to=end_date,sort_by='relevancy',language='en', page_size=100)
    
    articles = news_data['articles']
    sentiment_scores = []

    # You can use any sentiment analysis library to calculate sentiment scores
    # Here, we're just using the number of positive and negative words as an example
    positive_words = ['positive', 'good', 'growth', 'success']
    negative_words = ['negative', 'bad', 'decline', 'loss']

    for article in articles:
        content = article['title'] + ' ' + article['description']
        content = content.lower()
        pos_count = sum([content.count(word) for word in positive_words])
        neg_count = sum([content.count(word) for word in negative_words])
        sentiment_scores.append(pos_count - neg_count)

    return sentiment_scores

# Fetch fundamental data and news sentiment data for each ticker
fundamentals_data = []
news_sentiment_data = []

start_date = '2023-02-27'
end_date = '2023-03-25'

for ticker in tickers:
    print(f"Fetching fundamental data for {ticker}...")
    ticker_info = yf.Ticker(ticker).info
    fundamentals_data.append(ticker_info)

    sentiment_scores = fetch_news_sentiment(ticker, start_date, end_date)
    news_sentiment_data.append({'ticker': ticker, 'sentiment_scores': sentiment_scores})

print("Data collection complete.")
```

```python
import yfinance as yf
import pandas as pd
import datetime as dt
from GoogleNews import GoogleNews
from newsplease import NewsPlease

# Load the list of selected stocks
print("Loading selected stocks...")
selected_tickers_df = pd.read_csv("selected_stocks_financials.csv")
tickers = selected_tickers_df['Symbol'].tolist()

# Function to fetch news sentiment data
def fetch_news_sentiment(ticker, start_date, end_date):
    print(f"Fetching news sentiment data for {ticker}...")
    
    # Set up Google News
    googlenews = GoogleNews()
    googlenews.set_lang('en')
    googlenews.set_period('d')  # Can be changed to 'm' for month or 'y' for year
    googlenews.set_time_range(start_date, end_date)
    googlenews.set_encode('utf-8')

    # Search for news articles
    googlenews.search(ticker)

    # Extract news article URLs
    urls = [result['link'] for result in googlenews.results()]
    
    # Use news-please to fetch article content
    articles = []
    for url in urls:
        try:
            article = NewsPlease.from_url(url)
            articles.append(article)
        except Exception as e:
            print(f"Error fetching article: {e}")

    return articles

# Fetch fundamental data and news sentiment data for each ticker
fundamentals_data = []
news_sentiment_data = []

# Set the start and end dates
start_date = '2022-01-01'
end_date = '2023-03-25'

for ticker in tickers:
    print(f"Fetching fundamental data for {ticker}...")
    ticker_info = yf.Ticker(ticker).info
    fundamentals_data.append(ticker_info)

    articles = fetch_news_sentiment(ticker, start_date, end_date)
    news_sentiment_data.append({'ticker': ticker, 'articles': articles})

print("Data collection complete.")
```
