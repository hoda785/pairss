```python
import pandas as pd
import numpy as np

# Load the selected stocks financials and sentiment data from the CSV file
selected_stocks_df = pd.read_csv('selected_stocks_financials_sentiment.csv', index_col='symbol')

# Extract the sentiment scores as a separate Series
sentiment_scores = selected_stocks_df['sentiment_score']



price_data = pd.read_csv('hist adjust close financials.csv', index_col='Date', parse_dates=True)

# Extract the price data and sentiment data from the merged DataFrame
#price_data = selected_stocks_df.filter(regex='price_*')
sentiment_data = selected_stocks_df.filter(regex='sentiment_*')

# Calculate the correlation matrix for the price data
price_correlation_matrix = price_data.corr()
price_correlation_matrix
```

```python
adjusted_correlation_matrix_df = price_correlation_matrix.mul(sentiment_scores, axis=0).mul(sentiment_scores, axis=1)
adjusted_correlation_matrix_df
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Calculate the adjusted correlation matrix
price_data = pd.read_csv('hist adjust close financials.csv', index_col='Date', parse_dates=True)
price_returns = price_data.pct_change().dropna()

# Cluster the stocks based on the adjusted correlation matrix using the k-means algorithm
n_clusters = 4  # Set the number of clusters
kmeans = KMeans(n_clusters=n_clusters)
cluster_labels = kmeans.fit_predict(adjusted_correlation_matrix_df)

# Create a dictionary mapping stock symbols to cluster labels
symbol_cluster_dict = {symbol: cluster_labels[i] for i, symbol in enumerate(price_returns.columns)}

# Visualize the clustering results
fig, ax = plt.subplots(figsize=(18, 12))

# Set point size and color based on cluster labels
point_size = 150
colors = ['red', 'blue', 'green', 'orange']
for i in range(n_clusters):
    ax.scatter(price_returns.mean()[cluster_labels==i], price_returns.std()[cluster_labels==i], 
               s=point_size, color=colors[i], alpha=0.7, label=f'Cluster {i+1}')
    
    # Add stock labels to each point
    for symbol in price_returns.columns[cluster_labels==i]:
        ax.annotate(symbol, xy=(price_returns.mean()[symbol], price_returns.std()[symbol]), 
                    xytext=(5,-5), textcoords='offset points', fontsize=12)
        
ax.set_xlabel('Expected Returns', fontsize=16)
ax.set_ylabel('Risk', fontsize=16)
#ax.set_title(f'{n_clusters}-Cluster Solution', fontsize=20)
ax.grid(True, linestyle='--', linewidth=0.5)
ax.legend(fontsize=14)
plt.show()
```

```python
# Create a dictionary to store the tickers for each cluster
cluster_dict = {}
for i in range(n_clusters):
    cluster_dict[f'Cluster {i+1}'] = price_returns.columns[cluster_labels==i].tolist()

# Create a DataFrame from the cluster dictionary
cluster_dict
```

```python
# Create an empty DataFrame to store the stock tickers for each cluster
cluster_df = pd.DataFrame(columns=[f'Cluster {i+1}' for i in range(n_clusters)])

# Loop through the symbol_cluster_dict and assign each stock to its corresponding cluster in the DataFrame
for symbol, cluster in symbol_cluster_dict.items():
    cluster_df.loc[symbol] = [0] * n_clusters
    cluster_df.loc[symbol, f'Cluster {cluster+1}'] = 1

# Print the resulting DataFrame
print(cluster_df)
```

```python
cluster_1 = cluster_df[cluster_df['Cluster 1'] == 1].index.tolist()
cluster_2 = cluster_df[cluster_df['Cluster 2'] == 1].index.tolist()
cluster_3 = cluster_df[cluster_df['Cluster 3'] == 1].index.tolist()
cluster_4 = cluster_df[cluster_df['Cluster 4'] == 1].index.tolist()

print("Cluster 1:", cluster_1)
print("Cluster 2:", cluster_2)
print("Cluster 3:", cluster_3)
print("Cluster 4:", cluster_4)
```

```python
# Filter the DataFrame based on the index values and save to CSV
cluster_df.loc[cluster_1].to_csv('Cluster_1.csv', index=False)
cluster_df.loc[cluster_2].to_csv('Cluster_2.csv', index=False)
cluster_df.loc[cluster_3].to_csv('Cluster_3.csv', index=False)
cluster_df.loc[cluster_4].to_csv('Cluster_4.csv', index=False)
```
